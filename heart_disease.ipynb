```python
import pandas as pd
df=pd.read_csv("framingham_heart_disease.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['education'],axis =1)
df.head()
```

```python
df['cigsPerDay'].mean()
```

```python
cig_mean=round(df['cigsPerDay'].mean())
cig_mean
```

```python
df['cigsPerDay'].fillna(cig_mean,inplace = True)
```

```python
df['BPMeds'].mean()
```

```python
bp_mean=round(df['BPMeds'].mean())
bp_mean
```

```python
df['BPMeds'].fillna(cig_mean,inplace=True)
```

```python
df['BMI'].mean()
```

```python
bp_mean=round(df['BMI'].mean())
bp_mean
```

```python
df['BMI'].fillna(cig_mean,inplace=True)
```

```python
df['heartRate'].mean()
```

```python
bp_mean=round(df['heartRate'].mean())
bp_mean
```

```python
df['heartRate'].fillna(cig_mean,inplace=True)
```

```python
df['glucose'].mean()
```

```python
bp_mean=round(df['glucose'].mean())
bp_mean
```

```python
df['glucose'].fillna(cig_mean,inplace=True)
```

```python
df['totChol'].mean()
```

```python
bp_mean=round(df['totChol'].mean())
bp_mean
```

```python
df['totChol'].fillna(cig_mean,inplace=True)
```

```python
df.isnull().sum()
```

```python
x =df.drop('TenYearCHD',axis=1)
x
```

```python
y=df['TenYearCHD']
y
```

```python
from sklearn.model_selection import train_test_split
print(train_test_split(x,y,test_size=0.25))
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
x_train_predicted = classifier.predict(x_train)
x_train_predicted
```

```python
train_accuracy = accuracy_score(y_train,x_train_predicted)
train_accuracy
```

```python
x_test_prediction = classifier.predict(x_test)
x_test_prediction
```

```python
test_accuracy = accuracy_score(y_test,x_test_prediction)
test_accuracy
```

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test,x_test_prediction)
print(cr)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,x_test_prediction)
cm
```
